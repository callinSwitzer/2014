#### Problem 1(b)

Summarize the Salaries DataFrame to show the total salaries for each team for each year. Show the head of the new summarized DataFrame. 

```python
#your code here
grouped = sal.groupby(["teamID", "yearID"], as_index=False)
salByTeam = grouped.aggregate(np.sum)
salByTeam[0:5]
```

#### Problem 1(c)

Merge the new summarized Salaries DataFrame and Teams DataFrame together to create a new DataFrame
showing wins and total salaries for each team for each year year. Show the head of the new merged DataFrame.

**Hint**: Merge the DataFrames using `teamID` and `yearID`.

```python
wl = teams[["yearID", "franchID", "W", "L"]]
wl.columns = ["yearID", "teamID", "W", "L"]
merger = pd.merge(salByTeam, wl)
merger[0:5]
```

#### Problem 1(d)

How would you graphically display the relationship between total wins and total salaries for a given year? What kind of plot would be best? Choose a plot to show this relationship and specifically annotate the Oakland baseball team on the on the plot. Show this plot across multiple years. In which years can you detect a competitive advantage from the Oakland baseball team of using data science? When did this end? 

**Hints**: Use a `for` loop to consider multiple years. Use the `teamID` (three letter representation of the team name) to save space on the plot.  

### *I would graphically display the relationship with scatterplots. I can detect a competitive advantage from 2000 until 2006.  The advantage looks like it ended in 2007.*

```python
ss = set(merger.yearID)
for i in ss:

    yr = merger[(merger.yearID == i)] # getting a subset of the data frame for only a certain year
    yr.salary = yr.salary / 1000000 # changing salary to millions of dollars
    ok = yr[(yr.teamID == "OAK")] # the Oakland point
    
    # fitting a linear regression model
    model = sm.ols(formula='W ~ salary', data=yr)
    fitted = model.fit()


    #set figure and axis
    fig = plt.figure(figsize=(6, 4))

    #hide top and right axis
    ax = Subplot(fig, 111)
    fig.add_subplot(ax)

    #plot
    ax.plot(yr.salary, fitted.fittedvalues, '0.09') # plot a line
    ax.scatter(yr.salary, yr.W, s=50, facecolor='0.5', lw = 0) # plot each team
    ax.plot(ok.salary, ok.W, "o" ) # plot the Oakland point

    plt.xlabel("Salary (millions of dollars)") # xlabel
    plt.ylabel("Number of wins per season") # ylabel
    plt.title("Wins vs salary in " + str(i)) # title, according to year


    #do not display top and right axes
    ax.axis["right"].set_visible(False)
    ax.axis["top"].set_visible(False)
    ax.annotate("OAK", (ok.salary-3,ok.W + 3)) # annotate the Oakland point
 
```
