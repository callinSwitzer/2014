#### Problem 1(c)

Now, consider only team/season combinations in which the teams played 162 Games. Exclude all data from before 1947. Compute the per plate appearance rates for singles, doubles, triples, HR, and BB. Create a new pandas DataFrame called `stats` that has the teamID, yearID, wins and these rates.

**Hint**: Singles are hits that are not doubles, triples, nor HR. Plate appearances are base on balls plus at bats.

```python
### Your code here ###
t1 = teams.loc[teams.yearID > 1947,]
t2 = t1.loc[t1.G == 162,]
stat1 = t2[["teamID", "yearID","W", "AB", "H","2B","3B", "HR", "BB"]]
stat1["1BRate"] = (stat1["H"] - stat1["2B"]- stat1["3B"]- stat1["HR"])/ (stat1.AB + stat1.BB)
stat1["2BRate"] = stat1["2B"]/(stat1.AB + stat1.BB)
stat1["3BRate"] = stat1["3B"]/(stat1.AB + stat1.BB)
stat1["HRRate"] = stat1["HR"] / (stat1.AB + stat1.BB)
stat1["BBRate"] = stat1["BB"] / (stat1.AB + stat1.BB)

stats = stat1[["teamID", "yearID", "W", "1BRate", "2BRate", "3BRate", "HRRate", "BBRate"]]
stats.head()
```

#### Problem 1(f)

Build a simple linear regression model to predict the number of wins from the average adjusted singles, double, triples, HR, and BB rates. To decide which of these terms to include fit the model to data from 2002 and compute the average squared residuals from predictions to years past 2012. Use the fitted model to define a new sabermetric summary: offensive predicted wins (OPW). Hint: the new summary should be a linear combination of one to five of the five rates.


```python
## make model for before 2002, not just for 2002. 
## 2002
best = [None]*6

for ww in range(6):
    
    max1 = [None] * 100
    numToSample = ww
    items = ["1BRate", "2BRate", "3BRate", "HRRate", "BBRate"]
    

    for hh in range(len(max1)):
        items1 = random.sample(items, numToSample)
        s2002 = stat3.loc[stat3.yearID <= 2002]
        X = s2002[items1]
        y = s2002['W']

        ## fit a OLS model with intercept on "1BRate", "2BRate", "3BRate", "HRRate", "BBRate"
        X = sm.add_constant(X)
        est = sm.OLS(y, X).fit()
        res = [None]*len(range(2002, 2013))
        ctr = 0
        for w in range(2002, 2013):
            Y = stat3.loc[stat3.yearID == w][items1]
            Y = sm.add_constant(Y)
            Y["predictWin"] = est.predict(Y) # this is Offensive predicted wins
            Y["actualWin"] = stat3.loc[stat3.yearID == w]["W"]
            res[ctr] = sum((Y.actualWin - Y.predictWin)**2)/len(Y.actualWin)
            ctr = ctr + 1
        max1[hh] = (np.array(res).sum(), items1)
    mm = pd.DataFrame(max1)
    ff = mm.loc[mm[0] == mm[0].min()]
    #print((ff.iloc[0], max1[0]))
    best[ww] = (ff.iloc[0])

bb = pd.DataFrame(best)
bb.columns = ["residuals", "Predictors"]
bb
```

I fit a models for every linear combination of rates -- Singles, Doubles, Triples, Home Runs, and Base on Balls -- for the year 2002. The linear combination that gave the smallest sum of squared residuals was this model:  Wins = B1 + B2*(Base on balls rate) + B3*(Triple rate) + B4*(Home run rate) + random error.

Now we will create a similar database for individual players. Consider only player/year combinations in which the player had at least 500 plate appearances. Consider only the years we considered for the calculations above (after 1947 and seasons with 162 games). For each player/year compute singles, doubles, triples, HR, BB per plate appearance rates. Create a new pandas DataFrame called `playerstats` that has the playerID, yearID and the rates of these stats.  Remove the avareage for each year as for these rates as done in Problem 1(e). 

```python
### Your code here ###
#pd.set_option('display.height', 500)
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 200)
yearsWant = yearlyMean.yearID ## these are the years we want
players1 = players.loc[players.yearID.isin(yearsWant)]# cut data frame to only the years I want

## get total plate appearances for players
grouped = players1.groupby('playerID', as_index = False)
playerSums = grouped.sum()
playerSums.head()

## filter to players who have had more than 500 at-bats
player500 = playerSums.loc[(playerSums.AB + playerSums.BB) >= 500] # I'm ignoring sac bunts
playersWant = player500.playerID  #players who have had more than 500 at-bats


## filter players1 to only the players that I want
players2 = players1.loc[players1.playerID.isin(playersWant)]

## compute rates per plate appearance
players2["PlateAppearances"] = players2.AB + players2.BB
players2["2BRate"] = players["2B"] / (players2.PlateAppearances)
players2["3BRate"] = players["3B"] / (players2.PlateAppearances)
players2["BBRate"] = players["BB"] / (players2.PlateAppearances)
players2["HRRate"] = players["HR"] / (players2.PlateAppearances)
players2["1BRate"] = (players["H"] -players["2B"]- players["3B"]-players["HR"]) / (players2.PlateAppearances)
players3 = players2[["playerID", "yearID", "1BRate", "2BRate", "3BRate", "HRRate", "BBRate"]]

## compute yearly averages
grouped = players3.groupby('yearID', as_index = False)
yearM = grouped.mean()

## subtract yearly averages
players4 = players3.copy()
playerstats = players3.copy()
for i in yearsWant:
    smS = players4.loc[players4.yearID == i].copy()
    for j in ["1BRate", "2BRate", "3BRate", "HRRate", "BBRate"]:
        smS[j] = smS[j] - float(yearM[j][yearM.yearID == i])
    playerstats.loc[playerstats.yearID == i] = smS.copy()
```

```python
def accuracy_for_k_val_kFold(k, x_train, x_test, y_train, y_test, random_state=99):
    knn = sklearn.neighbors.KNeighborsClassifier(n_neighbors=k, 
                                             weights = 'uniform')
    knn.fit(x_train,y_train)
    y_hat = knn.predict(x_test)
    return sklearn.metrics.accuracy_score(y_test, y_hat, normalize=True)
```
